```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
top_50 = pd.read_csv("top50.csv")
```

```python
top_50.tail()
```

```python
top_50.drop("Unnamed: 0", axis = 1 , inplace = True)
```

```python
plt.figure(figsize=(15,8))
plt.title("Daily Global Popular Spotify songs in 2018-2019")
sns.lineplot(data = top_50)
```

```python
list(top_50.columns)
```

```python
plt.figure(figsize=(14,8))
sns.lineplot(data = top_50['Beats.Per.Minute'], label = 'BPM')
sns.lineplot(data = top_50['Popularity'], label = 'popularity')
plt.xlabel("DATA")
```
